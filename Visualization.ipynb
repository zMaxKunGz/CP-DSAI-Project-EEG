### Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Define csv path

```python
# SVM model
path_PCA = f'Result/testAccuracyPCA.csv'
path_noPCA = f'Result/testAccuracyNoPCA.csv'
```

```python
PCA = pd.read_csv(path_PCA)
noPCA = pd.read_csv(path_noPCA)
```

### Show result

```python
PCA.head()
```

```python
noPCA.head()
```

```python
def dfDescript(x):
    return pd.DataFrame(index=['min','max', 'mean'], data=[x.min(), x.max(), x.mean()])
```

```python
print("no-PCA results")
dfDescript(noPCA.drop(columns=['subject']))
```

```python
print("with PCA results")
dfDescript(PCA.drop(columns=['subject']))
```

### Plot graphs

```python
plt.plot(PCA['subject'],PCA['accSVM'], '.-', label="SVM")
plt.plot(PCA['subject'],PCA['acc3DCNN'], '.-', label="3D-CNN")
plt.plot(PCA['subject'],PCA['accParallel'], '.-', label="2D-Parallel-CNN")
plt.plot(PCA['subject'],PCA['accCMCNN'], '.-', label="CM-CW-CNN")
plt.ylim(50, 100)
plt.title("Accuracy: Model with PCA")
plt.xlabel("subject")
plt.ylabel("accuracy (%)")
plt.legend()
plt.show()
```

```python
plt.plot(noPCA['subject'],noPCA['accSVM'], '.-', label="SVM")
plt.plot(noPCA['subject'],noPCA['acc3DCNN'], '.-', label="3D-CNN")
plt.plot(noPCA['subject'],noPCA['accParallel'], '.-', label="2D-Parallel-CNN")
plt.plot(noPCA['subject'],noPCA['accCMCNN'], '.-', label="CM-CW-CNN")
plt.ylim(50, 100)
plt.title("Accuracy: Model without PCA")
plt.xlabel("subject")
plt.ylabel("accuracy (%)")
plt.legend()
plt.show()
```

```python
plt.plot(noPCA['subject'],noPCA['accParallel'], '.-', label="no PCA")
plt.plot(PCA['subject'],PCA['accParallel'], '.-', label="PCA")
plt.ylim(50, 100)
plt.title("Accuracy: Model Parallel-CNN")
plt.xlabel("subject")
plt.ylabel("accuracy (%)")
plt.legend()
plt.show()
```

```python
plt.plot(noPCA['subject'],noPCA['acc3DCNN'], '.-', label="no PCA")
plt.plot(PCA['subject'],PCA['acc3DCNN'], '.-', label="PCA")
plt.ylim(50, 100)
plt.title("Accuracy: Model 3D-CNN")
plt.xlabel("subject")
plt.ylabel("accuracy (%)")
plt.legend()
plt.show()
```

```python
plt.plot(noPCA['subject'],noPCA['accCMCNN'], '.-', label="no PCA")
plt.plot(PCA['subject'],PCA['accCMCNN'], '.-', label="PCA")
plt.ylim(50, 100)
plt.title("Accuracy: Model CM-CW-CNN")
plt.xlabel("subject")
plt.ylabel("accuracy (%)")
plt.legend()
plt.show()
```
